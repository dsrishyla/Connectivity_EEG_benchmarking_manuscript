```python
import glob
import xarray as xr
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy
```

```python
methods = ['coh','ciplv','imcoh','wpli2']
```

```python
#Functions

def outlier_indices(data):
    q1 = np.percentile(data, 25, method='midpoint')
    q3 = np.percentile(data, 75, method='midpoint')
    iqr = q3 - q1
    upper = q3 + 1.5 * iqr
    lower = q1 - 1.5 * iqr
    upper_outliers = np.where(data >= upper)[0]
    lower_outliers = np.where(data <= lower)[0]
    outlier_index_list = list(upper_outliers) + list(lower_outliers)
    
    return outlier_index_list

def remove_outliers(data):
    outliers_removed = []
    
    for i in range(0,len(data)):
        if i not in outlier_indices(data):
            outliers_removed.append(data[i])
            
    return outliers_removed

def bootstrap_ci(group1, group2):
    
    ####################################################
    #sample from pooled 1000 times
    n1 = len(group1)
    n2 = len(group2)
    
    bootstrap_sample1 = []
    bootstrap_sample2 = []
    
    for i in range(1000):
        sample1 = []
        sample2 = []
        index1 = np.random.choice(range(0,n1),n1,replace=True)
        index2 = np.random.choice(range(0,n2),n2,replace=True)
        sample1 = np.array(group1)[index1]
        sample2 = np.array(group2)[index2]
        bootstrap_sample1.append(sample1)
        bootstrap_sample2.append(sample2)
        
    diff_list = []
    for i in range(1000):
        mean1 = np.mean(bootstrap_sample1[i])
        mean2 = np.mean(bootstrap_sample2[i])
    
        diff = mean2 - mean1
        diff_list.append(diff)
    

    diff_list.sort()


    perc_025 = diff_list[24]
    perc_975 = diff_list[974]

    
    return perc_025, perc_975

def confint_95_single(data):

    bootstrap_samples = []

    for i in range(1000):
    #create 1000 bootstrap samples
        n = len(data)
        index = np.random.choice(range(0,n),n,replace=True)
        sample = np.array(data)[index]
        bootstrap_samples.append(sample)
    
    #calculate 1000 means
    means = []
    for sample in bootstrap_samples:
        mean = np.mean(sample)
        means.append(mean)

    #sort means list
    means.sort()
    
    #choose 25th value, choose 975th value
    perc_025 = means[24]
    perc_975 = means[974]

    
    return perc_025, perc_975
```

```python
ec_means = {}
eo_means = {}
ec_stds = {}
eo_stds = {}
d_vals = {}
p_vals = {}

mean_lists_ec = {}
mean_lists_eo = {}

vn = ['fusiform-rh', 'fusiform-lh', 'lingual-rh', 'lingual-lh', 'cuneus-rh', 'cuneus-lh',
                      'lateraloccipital-rh', 'lateraloccipital-lh']

xarrays = glob.glob(f'/Users/lina_01/Desktop/mpi_lemon/undirected_outputs/coh/*_coh_EC.nc')
subjects = []
for array in xarrays:
    subject = array.split('/')[7].split('_')[0]
    subjects.append(subject)

for method in methods:
    xarrays = glob.glob(f'/Users/lina_01/Desktop/mpi_lemon/undirected_outputs/{method}/*_{method}_EC.nc')
                
    xarrays_combined = []
    for array in xarrays:
        xarrays_combined.append(xr.open_dataarray(array))
    
    xarrays_combined = xr.concat(xarrays_combined, pd.Index(subjects, name="subjects"))
    xarrays_combined_vn = xarrays_combined.sel(region1 = vn, region2 = vn)

    mean_ec = []
    
    for subject in subjects:
        sub_data = xarrays_combined_vn.sel(subjects=subject)
        mean = sub_data.mean(dim=["bootstrap_samples","region1","region2","frequencies"]).values
        mean_ec.append(float(mean))

    mean_lists_ec[method] = mean_ec

    overall_mean_ec = np.mean(mean_ec)
    ec_means[method] = overall_mean_ec

    overall_std_ec = np.std(mean_ec)
    ec_stds[method] = overall_std_ec


for method in methods:
    xarrays = glob.glob(f'/Users/lina_01/Desktop/mpi_lemon/undirected_outputs/{method}/*_{method}_EO.nc')
                
    xarrays_combined = []
    for array in xarrays:
        xarrays_combined.append(xr.open_dataarray(array))
    
    xarrays_combined = xr.concat(xarrays_combined, pd.Index(subjects, name="subjects"))
    xarrays_combined_vn = xarrays_combined.sel(region1 = vn, region2 = vn)

    mean_eo = []
    
    for subject in subjects:
        sub_data = xarrays_combined_vn.sel(subjects=subject)
        mean = sub_data.mean(dim=["bootstrap_samples","region1","region2","frequencies"]).values
        mean_eo.append(float(mean))
    
    mean_lists_eo[method] = mean_eo
    
    overall_mean_eo = np.mean(mean_eo)
    eo_means[method] = overall_mean_eo
    
    overall_std_eo = np.std(mean_eo)
    eo_stds[method] = overall_std_eo

for method in methods:

    wilcox, p = scipy.stats.wilcoxon(mean_lists_ec[method], mean_lists_eo[method])
    p_vals[method] = p
    
    pooled_sd = np.sqrt((((len(mean_lists_ec[method])-1)*(ec_stds[method]**2) + (len(mean_lists_eo[method])-1)*(eo_stds[method])**2))/(len(mean_lists_ec[method]) + len(mean_lists_eo[method]) - 2))
    cohens_d = (ec_means[method]- eo_means[method])/pooled_sd
    d_vals[method] = cohens_d
```

```python
ci_dict = {}

for method in methods:
    ci_dict[method] = bootstrap_ci(mean_lists_ec[method], mean_lists_eo[method])
```

```python
graph_ci_eo_dict = {}

for method in methods:
    lower_limit = confint_95_single(mean_lists_eo[method])[0]
    upper_limit = confint_95_single(mean_lists_eo[method])[1]
    err = upper_limit - lower_limit
    graph_ci_eo_dict[method] = err

graph_ci_ec_dict = {}

for method in methods:
    lower_limit = confint_95_single(mean_lists_ec[method])[0]
    upper_limit = confint_95_single(mean_lists_ec[method])[1]
    err = upper_limit - lower_limit
    graph_ci_ec_dict[method] = err
```
