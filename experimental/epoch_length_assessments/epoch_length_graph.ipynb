```python
methods = ['coh','ciplv','imcoh','dwpli','gpdc','psgp','ddtf']
```

```python
new_val_dict = {}

for method in methods:
    mean_df_subset = mean_df[mean_df['metric']==method]
    mean_list = list(mean_df_subset['mean_conn'])
    new_vals = []
    for val in mean_list:
        new = (val - np.min(mean_list))/(np.max(mean_list) - np.min(mean_list))
        new_vals.append(new)
        
    new_val_dict[method] = new_vals
    
```
